# Whisper 개요
- 개요: OpenAI에서 공개한 음성 인식(STT) 모델
- 제공 형태
    - 클라우드 API: 분당 $0.06 (2025년 4월 기준)
    - 오픈웨이트 모델: MIT 라이선스, 로컬 실행 가능
- 주요 기능
    - 전사(Transcription): 음성을 동일 언어 텍스트로 변환
    - 번역(Translation): 모든 지원 언어 음성을 영어 텍스트로 변환
- 다국어 지원
    - 영어, 한국어 포함 98개 이상 언어 지원
- 입력 파일 제한
    - 형식: mp3, wav, mp4 등
    - 크기: 최대 25MB

## Whisper API 실습: 전사(Transcriptions)

```python
from openai import OpenAI
from dotenv import load_dotenv
import os

# .env에서 API 키 로드
load_dotenv()
api_key = os.getenv('OPEN_API_KEY')
# OpenAI 클라이언트 초기화
client = OpenAI(api_key=api_key)

# 전사할 오디오 파일 경로
audio_file_path = '../audio/lsy_audio_2023_58s.mp3'

# Whisper 전사 호출
with open(audio_file_path, 'rb') as audio_file:
    transcription = client.audio.transcriptions.create(
        model='whisper-1',
        file=audio_file
    )

transcription
```

## Whisper API 실습: 번역(Translations)

```python
# 번역(Transcribe + Translate) API 호출
with open(audio_file_path, 'rb') as audio_file:
    translations = client.audio.translations.create(
        model="whisper-1",
        file=audio_file,
    )

translations
```
