## 05-1 음성을 텍스트로 변환하기

- STT(Speech-To-Text)로 음성에서 텍스트를 생성
- 회의록 자동화의 핵심 입력 단계


### 실습. OpenAI의 Whisper API 활용하기

- 한국어 포함 90~100개 언어 지원
- 별도 LLM 없이 번역(주로 영어) 제공
- 비용: 분당 6센트(2025년 4월 기준)


#### OpenAI API 클라이언트 선언

- API 키 설정과 클라이언트 초기화


```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv('OPEN_API_KEY')
client = OpenAI(api_key=api_key)
```

#### MP3의 음성을 텍스트로 변환

- 음성 파일 업로드 후 텍스트 추출


```python
audio_file_path = '../audio/lsy_audio_2023_58s.mp3'  # MP3 파일 경로 입력

with open(audio_file_path, 'rb') as audio_file:
    transcription = client.audio.transcriptions.create(
        model='whisper-1',
        file=audio_file
    )

transcription
```

#### MP3의 한국어 음성을 영어로 번역하기

- 동일한 입력으로 번역 결과 생성


```python
# transcriptions -> translations
with open(audio_file_path, 'rb') as audio_file:
    transcription = client.audio.translations.create(
        model="whisper-1",
        file=audio_file,
    )

transcription
```
